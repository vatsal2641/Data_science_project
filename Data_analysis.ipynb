```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/EducationDataset_2023-24.csv")
```

```python
df.head()
```

```python
columns = df.columns.tolist()
columns
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df[" PASS PERCENTAGE IN CLASS X - \n(Before Compt.) - 2023-24"],
            df[" QUALITY INDEX IN CLASS X - \n(Before Compt) - 2023-24"],
            color='royalblue', s=80)
plt.title("Relationship between Pass % and Quality Index (Class X)")
plt.xlabel("Pass Percentage (Class X)")
plt.ylabel("Quality Index (Class X)")
plt.grid(True)
plt.savefig("Scatter Plot")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.boxplot([df[" PASS PERCENTAGE IN CLASS X - \n(Before Compt.) - 2023-24"],
             df["PASS PERCENTAGE IN CLASS XII - (Before Compt.) - 2023-24"]],
            labels=["Class X", "Class XII"],
            patch_artist=True, boxprops=dict(facecolor="lightgreen"))
plt.title("Distribution of Pass Percentages")
plt.ylabel("Pass Percentage (%)")
plt.grid(True, axis='y')
plt.savefig("Boxplot")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(df["District"],
        df["PASS PERCENTAGE IN CLASS XII - (Before Compt.) - 2023-24"],
        color='coral')
plt.xticks(rotation=45, ha='right')
plt.title("Class XII Pass Percentage by District")
plt.xlabel("District")
plt.ylabel("Pass Percentage (%)")
plt.savefig("Barplot")
plt.show()
```
